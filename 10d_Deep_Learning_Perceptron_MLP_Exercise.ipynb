# Wielowarstwowy Perceptron (MLP) – klasyfikacja cukrzycy (Pima Indians Diabetes)

&#x20;


W tym projekcie zbadamy krok po kroku zastosowanie sieci neuronowej typu **Multi-Layer Perceptron (MLP)** do zadania klasyfikacji cukrzycy na podstawie danych medycznych pacjentek z plemienia Pima (zbiór **Pima Indians Diabetes**). Zestaw danych pochodzi z National Institute of Diabetes and Digestive and Kidney Diseases i zawiera dane diagnostyczne pozwalające przewidywać wystąpienie cukrzycy u kobiet pochodzenia Pima Indian w ciągu 5 lat Mamy 768 obserwacji oraz 8 cech wejściowych i 1 zmienną wyjściową (klasę), przy czym liczba przypadków cukrzycy (klasa 1) jest nieco mniejsza od liczby przypadków bez cukrzycy (klasa 0) – dane są umiarkowanie niezbalansowane

**Cechy (zmienne niezależne):**

* **Pregnancies** – liczba przebytych ciąż.
* **Glucose** – stężenie glukozy we krwi po 2 godzinach testu tolerancji glukozy.
* **BloodPressure** – rozkurczowe ciśnienie krwi (mm Hg).
* **SkinThickness** – grubość fałdu skórnego na tricepsie (mm).
* **Insulin** – poziom insuliny w surowicy po 2 godzinach (mu U/ml).
* **BMI** – indeks masy ciała (kg/m²).
* **DiabetesPedigreeFunction** – funkcja dziedziczności cukrzycy (uwarunkowania genetyczne).
* **Age** – wiek (lata).

**Zmienna docelowa (klasa):**

* **Outcome** – wynik diagnostyczny (0 = brak cukrzycy, 1 = cukrzyca)
